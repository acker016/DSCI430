## Using Pandas Chucksize to read in and inspect the columns

```python
input_files = [pd.read_csv('../MinneMUDACProject1/MinneMUDAC_raw_files/' + str(file) + '_metro_tax_parcels.txt', sep= '|', chunksize=500)
             for file in range(2002, 2016)]# if file != 2003]
```

```python
first_chunks = [first(file).columns for file in input_files]
first_chunks
```

```python
heading_sets = [set(file) for file in first_chunks]
heading_sets
```

```python
all_cols = set([])
for heading_set in heading_sets:
    all_cols = all_cols.union(heading_set)
#all_cols
```

```python
all_cols.intersection(first_chunks[0])
```

```python
all_cols.difference(first_chunks[0])
```

```python
common_items = all_cols
for all_cols in heading_sets:
    common_items = common_items.intersection(all_cols)
len(common_items)
```

```python
[s.difference(common_items) for s in heading_sets]
```

```python
files_zip_years = [(i, s.difference(common_items)) for i, s in zip(range(2002, 2016), heading_sets)]
files_zip_years
```
